```python
import time
import os
from PIL import Image
import sys

print("Process Start.")

start_time = time.time()

# 수정할 사진 폴더명 입력
directory = "./14_15_splash"

# 퍼센트 입력 (예: 0.5는 50%로 줄임)
percent = 0.5

# 결과 저장 폴더 생성
out_dir = "14_15_splash_resized"

if out_dir not in os.listdir():
    os.mkdir(out_dir)

input_files = os.listdir(directory)

for filename in input_files:
    # 파일 확장자 분리
    resultName, exp = os.path.splitext(filename)
    exp = exp.lower()
    print(filename + " => " + resultName)
    # 이미지 파일 확장자 아닌 파일이 있는경우 거름
    if exp not in [".jpg", ".jpeg", ".png", ".bmp"]:
        continue

    # 이미지를 불러옵니다.
    image = Image.open(os.path.join(directory, filename))

    # RGBA 모드 이미지 처리
    if image.mode == 'RGBA':
        image = image.convert('RGB')

    # 이미지의 크기를 알아냅니다.
    Xdim, Ydim = image.size

    # 퍼센트 만큼 사이즈 조절
    Xdim *= percent
    Ydim *= percent

    # 이미지 사이즈를 변경합니다.
    image = image.resize((int(Xdim), int(Ydim)))

    # 이미지 저장
    image.save(os.path.join(out_dir, resultName.lower() + ".webp"))

    image.close()

print("Process Done.")

end_time = time.time()
print("The Job Took " + str(end_time - start_time) + " seconds.")
```
